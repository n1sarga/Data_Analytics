```python
import pandas as pd

df_sal = pd.read_csv("Salary_Data.csv")

df_sal.head()
```

```python
df_sal['Education Level'].unique()
```

```python
df_sal["Education Level"] = df_sal["Education Level"].replace( "Bachelor's Degree", "Bachelor's")
df_sal["Education Level"] = df_sal["Education Level"].replace("Master's Degree", "Master's")
df_sal["Education Level"] = df_sal['Education Level'].replace('phD', "PhD")
df_sal["Education Level"] = df_sal['Education Level'].replace('phd', "PhD")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

x = df_sal[df_sal["Gender"]=="Male"].Salary.mean()
y = df_sal[df_sal["Gender"]=="Female"].Salary.mean()
z = df_sal[df_sal["Gender"]=="Other"].Salary.mean()

list = [x, y, z]

X_axis = ["Male", "Female", "Others"]
sns.barplot(x = X_axis, y = list, hue = X_axis, palette = "Reds")
plt.title("Bar Chart")
plt.xlabel("Gender")
plt.ylabel("Mean Salary")
plt.show()

print ("Male Average Income:", x, "Female Average Income:", y, "Other Average Income:", z)
```

```python
import pandas as pd
import seaborn as sns

male_salaries = df_sal[df_sal["Gender"] == "Male"].Salary

sns.histplot(male_salaries, bins = 45)

plt.show()

print ("Minimum income of male jobholders:", male_salaries.min())
print ("Maximum income of male jobholders:", male_salaries.max())
print ("Most earned amount:", male_salaries.mode())
```

```python
import pandas as pd
import seaborn as sns

female_salaries = df_sal[df_sal["Gender"] == "Female"].Salary

sns.histplot(female_salaries, bins = 35)

plt.show()
print ("Minimum income of female jobholders:", female_salaries.min())
print ("Maximum income of female jobholders:", female_salaries.max())
print ("Most earned amount:", female_salaries.mode())
```

```python
import pandas as pd
import seaborn as sns

other_salaries = df_sal[df_sal["Gender"] == "Other"].Salary

sns.histplot(other_salaries, bins = 35)

plt.show()
print ("Minimum income of other jobholders:", other_salaries.min())
print ("Maximum income of other jobholders:", other_salaries.max())
print ("Most earned amount:", other_salaries.mode())
```

```python
skewness = df_sal[df_sal["Gender"] == "Male"].Salary.skew()
kurtosis = df_sal[df_sal["Gender"] == "Male"].Salary.kurtosis()

sns.displot(df_sal[df_sal["Gender"] == "Male"].Salary, kde=True)
plt.title("Salary Distribution for Male Employees")
plt.xlabel("Salary")
plt.show()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
skewness = df_sal[df_sal["Gender"] == "Female"].Salary.skew()
kurtosis = df_sal[df_sal["Gender"] == "Female"].Salary.kurtosis()

sns.displot(df_sal[df_sal["Gender"] == "Female"].Salary, kde=True)
plt.title("Salary Distribution for Female Employees")
plt.xlabel("Salary")
plt.show()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
skewness = df_sal[df_sal["Gender"] == "Other"].Salary.skew()
kurtosis = df_sal[df_sal["Gender"] == "Other"].Salary.kurtosis()

sns.displot(df_sal[df_sal["Gender"] == "Other"].Salary, kde=True)
plt.title("Salary Distribution for Other Employees")
plt.xlabel("Salary")
plt.show()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```
